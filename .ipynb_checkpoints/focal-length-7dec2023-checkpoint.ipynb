```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as sp_optimize
%matplotlib inline
```

```python
csv_file_path = './data/data.csv'
df = pd.read_csv(csv_file_path)
train_data = df.head(4)
test_data = df.tail(1)
print(df)
```

```python
plt.scatter(train_data.WD, train_data.f)
plt.xlabel("Working Distance")
plt.ylabel("f (calculated)")
plt.title("Training data")
```

```python

# VV sample function provided by Will VV
initial_guess = [1.0, 0.01, 1.0]
def sample_function(WD, A, B, C):
    return A*np.exp(-B * WD) + C

train_params, train_covariance = sp_optimize.curve_fit(sample_function, train_data["WD"], train_data["f"], p0=initial_guess, maxfev=10000)

# print(my_params[0])
# print(my_covariance[0])

WD_fit = np.linspace(df["WD"].min(), df["WD"].max() * 1.2, 100)
f_fit = sample_function(WD_fit, *train_params)

plt.scatter(train_data["WD"], train_data["f"], label='Training Data')

plt.plot(WD_fit, f_fit, label='Fitted Curve', color='red')
plt.title("First four with curve")
plt.xlabel("f")
plt.ylabel("WD")

test_prediction = sample_function(test_data["WD"].values[0], *train_params)

print("Predicted value on test set:", test_prediction)
print("Actual value on test set:", test_data["f"].values[0])

# Highlight the training points
plt.scatter(train_data["WD"], train_data["f"], label='Training Points', color='green')

# Highlight the test point
plt.scatter(test_data["WD"], test_data["f"], label='Test Point', color='orange', marker='X')
```
